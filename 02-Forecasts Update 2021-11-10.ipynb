```python
#format for JSON
dfForecastJSON2024 = sdfNewForecastsFormatted[['SEGID','F2024','CO_FIPS']].copy()
dfForecastJSON2024['YEAR'] = 2024
dfForecastJSON2024.columns = ('S','F','CO_FIPS','Y')
dfForecastJSON2024 = dfForecastJSON2024[['S','Y','F','CO_FIPS']]
display(dfForecastJSON2024)

dfForecastJSON2030 = sdfNewForecastsFormatted[['SEGID','F2030','CO_FIPS']].copy()
dfForecastJSON2030['YEAR'] = 2030
dfForecastJSON2030.columns = ('S','F','CO_FIPS','Y')
dfForecastJSON2030 = dfForecastJSON2030[['S','Y','F','CO_FIPS']]
display(dfForecastJSON2030)

dfForecastJSON2040 = sdfNewForecastsFormatted[['SEGID','F2040','CO_FIPS']].copy()
dfForecastJSON2040['YEAR'] = 2040
dfForecastJSON2040.columns = ('S','F','CO_FIPS','Y')
dfForecastJSON2040 = dfForecastJSON2040[['S','Y','F','CO_FIPS']]
display(dfForecastJSON2040)

dfForecastJSON2050 = sdfNewForecastsFormatted[['SEGID','F2050','CO_FIPS']].copy()
dfForecastJSON2050['YEAR'] = 2050
dfForecastJSON2050.columns = ('S','F','CO_FIPS','Y')
dfForecastJSON2050 = dfForecastJSON2050[['S','Y','F','CO_FIPS']]
display(dfForecastJSON2050)

dfForecastJSON = pd.concat([dfForecastJSON2024,dfForecastJSON2030,dfForecastJSON2040,dfForecastJSON2050])
display(dfForecastJSON)

dfForecastJSON.to_json(os.path.join(dirWebApp, r'forecasts.json'),orient='records')
```
